```python
import numpy as np

currentDataClass = [1, 3, 3, 2, 5, 5, 3, 2, 1, 4, 3, 2, 1, 1, 2] # extract the different classes #actual
predictedClass =   [1, 2, 3, 4, 2, 3, 3, 2, 1, 2, 3, 1, 5, 1, 1] # initialize the confusion matrix with zeros #predicted

def confusion_matrix(actual, predicted):
    classes = np.unique(actual)
    matrix = np.zeros((len(classes), len(classes)))
    
    for i in range(len(classes)):
        for j in range(len(classes)):
            print('actual:')
            print(actual == classes[i])
            print('predicted')
            print (predicted == classes[j])
            matrix[i,j] = np.sum((actual == classes[i]) & (predicted == classes[j])) #np.sum adds all elements in the list
            
    return matrix

confusion_matrix(currentDataClass, predictedClass)
```

* actual == classes[i] returns a list of true and false if each element = classes[i] or not
* predicted == classes[i] returns a list of true and false if each elements = classes[i] or not
* & does bitwise AND - returns a list in which an element is true only if both of the conditions are true
* np.sum adds all elements in the list
